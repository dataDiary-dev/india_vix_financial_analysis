# 04 — Business Recommendations
## India VIX Analysis (2009–2021)

---

## Finding 1 — March 2020 was the Peak Panic Event
VIX hit 83.6 — the highest in the dataset — during COVID-19.
**Recommendation:** Historically, VIX spikes above 50 have been
followed by sharp recoveries. Investors who held through panic
zones saw the best 1-year returns.

---

## Finding 2 — Market spends 60%+ of time in Calm/Moderate zones
**Recommendation:** India is a structurally stable market.
Long term SIP investors should not react to short term VIX spikes.
